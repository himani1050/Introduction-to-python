NUMPY:
NumPy (Numerical Python) is the foundation for almost all numerical computing in Python. It provides high-performance, multi-dimensional array objects and tools for working with them.

```python
import numpy as np
arr1 = np.array([1, 2, 3, 4])
arr2 = np.arange(1, 10, 2) 
arr3 = np.arange(1, 10, 2).reshape(5, 1) # [1, 3, 5, 7, 9] as a column vector

print(arr1)
print(arr2)
print(arr3)

print(arr1 * 3)

# Aggregation
print(f"Mean: {arr1.mean()}")
print(f"Sum: {arr1.sum()}")
```

Matplotlib is the foundational plotting library, and Seaborn is built on top of it, providing a high-level interface for drawing attractive and informative statistical graphics.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data_vis = {'Score': [85, 90, 78, 92, 88, 75],
            'Hours_Studied': [4, 6, 3, 7, 5, 2]}
df_vis = pd.DataFrame(data_vis)

print("--- Data Visualization Example ---")

# Scatter plot of Score vs. Hours Studied
sns.scatterplot(x='Hours_Studied', y='Score', data=df_vis)
plt.title('Score vs. Hours Studied (Seaborn)')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.show() 

# Basic Line Plot (trend over time)
x = np.arange(5)
y = x**2
plt.figure() # Start a new plot
plt.plot(x, y, marker='o')
plt.title('Simple Matplotlib Line Plot')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file; setting the first column as the index
df_gene_expression = pd.read_csv('differential_expression_test.csv', index_col=0)

# Creating the heatmap figure
plt.figure(figsize=(7, 6)) 

sns.heatmap(df_gene_expression,
            cmap='viridis',      # Color map
            annot=True,          # Show the actual values in the cells
            fmt=".1f",           # Format annotations to one decimal place
            linewidths=.5,       # Add lines between cells
            cbar_kws={'label': 'Expression Level'}) # Label the color bar

plt.title('Differential Gene Expression Heatmap')
plt.xlabel('Experimental Condition')
plt.ylabel('Gene')
```

Biopython is a set of tools for biological computation, specifically designed for working with biological data formats, sequence analysis, and connecting to common bioinformatics databases.

```python
from Bio.Seq import Seq

#Create a DNA Sequence object
dna_seq = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG")

print(f"Original DNA: {dna_seq}")
print(f"Length: {len(dna_seq)}")

# DNA -> RNA (Transcription)
mrna_seq = dna_seq.transcribe()
print(f"mRNA (RNA): {mrna_seq}")

# RNA -> Protein (Translation)
protein_seq = mrna_seq.translate(to_stop=True)
print(f"Protein: {protein_seq}")
```
