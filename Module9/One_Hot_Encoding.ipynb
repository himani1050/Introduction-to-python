```python
#one hot encoding
data1 = pd.read_csv('employee.csv')
data1
```

```python
data1.dtypes
```

```python
from sklearn.preprocessing import OneHotEncoder
categeorical_columns = data1.select_dtypes(include = ['object']).columns.tolist()
encoder = OneHotEncoder(sparse_output = False)

one_hot_encoded = encoder.fit_transform(data1[categeorical_columns])

one_hot_df = pd.DataFrame(one_hot_encoded, columns = encoder.get_feature_names_out(categeorical_columns))
df_encoded = pd.concat([data1, one_hot_df], axis = 1)
```

```python
df_encoded
```

```python
df_encoded = df_encoded.drop(categeorical_columns, axis = 1)
```
